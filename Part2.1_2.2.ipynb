```python
import mysql.connector
import mysql.connector as mariadb

def get_db_connection():
    con = mariadb.connect(host='localhost', port= 3307,user="root",password="2011",database="creditcard_capstone")
    return con
```

```python
from getpass import getpass
```

```python
#2.1.1.Used to display the transactions made by customers living in a given zip code for a given month and year. Order by day in descending order.

def transactions_by_customer_zip_year_month():
 
        sql = """SELECT cc.* FROM cdw_sapp_credit_card1 cc, cdw_sapp_customer c WHERE cc.CUST_SSN = c.SSN AND c.CUST_ZIP = %s
        AND cc.TIMEID LIKE %s ORDER BY timeid ASC"""

        zip = input("Enter customer zip:\t").strip()
        year = input("Enter transaction year:\t").strip()
        month = input("Enter transaction month:\t").strip().zfill(2)
        

        db_con = get_db_connection()
        cursor = db_con.cursor()
        #cursor.execute(sql, (zip, year + '-' + month +'-%',))
        cursor.execute(sql, (zip, year+ month +'%',))
        records = cursor.fetchall()
        
        for row in records:
            print(row)
        db_con.close()
        print('Database operation completed successfully')

transactions_by_customer_zip_year_month()
```

```python
#2.1.2.Used to display the number and total values of transactions for a given type.


def transactions_value_count():
   
    sql = "SELECT COUNT(*), SUM(TRANSACTION_VALUE) FROM cdw_sapp_credit_card WHERE TRANSACTION_TYPE=%s"

    tn_type = input("Transaction Type:\t")
    
    db_con = get_db_connection()
    cursor = db_con.cursor()
    cursor.execute(sql,(tn_type,))
    records = cursor.fetchall()
    for row in records:
        print(row)
    db_con.close()
    print('Database operation completed successfully')
    
transactions_value_count() 
```

```python
#2.1.3.Used to display the number and total values of transactions for branches in a given state.

def transactions_value_state():
    sql = """SELECT COUNT(*), SUM(cc.TRANSACTION_VALUE) FROM cdw_sapp_credit_card cc, cdw_sapp_branch b  WHERE cc.BRANCH_CODE = b.BRANCH_CODE 
           AND b.BRANCH_STATE = %s
           GROUP BY b.BRANCH_CODE"""
    
    state = input("Enter State:\t")
  
    db_con = get_db_connection()
    cursor = db_con.cursor()
    cursor.execute(sql,(state,))
    records = cursor.fetchall()
    for row in records:
        print(row)
    db_con.close()
    print('Database operation completed successfully')

transactions_value_state()    
```

```python
 #2.2.1.Used to check the existing account details of a customer.
    

def customer_account_details():
   
    sql = "SELECT cc.* FROM cdw_sapp_credit_card cc, cdw_sapp_customer c  WHERE cc.CUST_SSN = c.SSN and c.FIRST_NAME= %s"

    customer = input("customer Firstname:\t")
    
    db_con = get_db_connection()
    cursor = db_con.cursor()
    cursor.execute(sql,(customer,))
    records = cursor.fetchall()
    for row in records:
        print(row)
    db_con.close()
    print('Database operation completed successfully')
    
customer_account_details()     
    
```

```python
#2.2.3. Used to generate a monthly bill for a credit card number for a given month and year.

def month_bill():
   
    sql = "SELECT sum(TRANSACTION_VALUE) FROM cdw_sapp_credit_card1 where CUST_CC_NO = %s and TIMEID LIKE %s"

    creditcard_no = input("Enter credit card no:\t")
    year = input("Enter transaction year:\t")
    month = input("Enter transaction month:\t").zfill(2)
    
    db_con = get_db_connection()
    cursor = db_con.cursor()
    cursor.execute(sql,(creditcard_no, year+ month +'%',))
    records = cursor.fetchall()
    print(records)
    db_con.close()
    print('Database operation completed successfully')
    
month_bill() 
```

```python
#2.2.4.Used to display the transactions made by a customer between two dates. Order by year, month, and day in descending order.

def tran_between_date():
   
    sql = """SELECT TRANSACTION_ID, TRANSACTION_VALUE,TIMEID FROM cdw_sapp_credit_card1 WHERE TIMEID between %s and %s 
          AND CUST_SSN = %s ORDER BY TIMEID DESC """

    
    date1 = input("Enter starting transaction date:\t")
    date2 = input("Enter ending transaction date:\t")
    ssn_no = getpass("Enter ssn no:\t")
    
    db_con = get_db_connection()
    cursor = db_con.cursor()
    cursor.execute(sql,(date1, date2, ssn_no,))
    records = cursor.fetchall()
    for row in records:
        print(row)
    db_con.close()
    print('Database operation completed successfully')
    
tran_between_date() 
```
